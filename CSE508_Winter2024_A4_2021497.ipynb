```python
import pandas as pd
import re
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

# Download NLTK resources
import nltk
nltk.download('punkt')
nltk.download('stopwords')

# Load DataFrame
# Assuming you already have df_selected from the previous steps

# Define preprocessing function
def preprocess_text(text):
    if pd.isna(text):
        return ""
    # Lowercase the text
    text = text.lower()

    # Remove special characters (excluding alphanumeric characters and spaces)
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)

    # Tokenize the text
    tokens = word_tokenize(text)

    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    tokens = [token for token in tokens if token not in stop_words and not token.isdigit()]

    # Join tokens back into a single string
    preprocessed_text = ' '.join(tokens)

    return preprocessed_text

# Create a new DataFrame for storing preprocessed text
df_preprocessed = pd.DataFrame(columns=['Summary', 'Review'])

# Apply preprocessing function to 'Summary' and 'Review' columns
df_preprocessed['Summary'] = df_selected['Summary'].apply(preprocess_text)

```

```python
from transformers import GPT2Tokenizer, Trainer, TrainingArguments
from torch.utils.data import Dataset

class TextDataset(Dataset):
    def __init__(self, encodings):
        self.encodings = encodings

    def __getitem__(self, idx):
        item = {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}
        item['labels'] = item['input_ids']
        return item

    def __len__(self):
        return len(self.encodings.input_ids)

class DataProcessor:
    def __init__(self, tokenizer):
        self.tokenizer = tokenizer

    def process_data(self, data_df, max_length=512):
        tokenized_data = self.tokenize_data(data_df, max_length=max_length)
        return TextDataset(tokenized_data)

    def tokenize_data(self, data_df, max_length=512):
        tokenized_data = self.tokenizer(data_df["Text"].tolist(), padding=True, truncation=True, max_length=max_length, return_tensors="pt")
        return tokenized_data

# Example usage
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
tokenizer.pad_token = tokenizer.eos_token
data_processor = DataProcessor(tokenizer)

# Assuming train_data is your DataFrame
# train_data = pd.read_csv("your_dataset.csv")  # Load your DataFrame here
train_data_tokenized = data_processor.process_data(train_data)

# Initialize the training dataset
eval_data_tokenized = data_processor.process_data(test_data)

# Initialize the training arguments
training_args = TrainingArguments(
    output_dir='./results',          # output directory
    num_train_epochs=3,              # total number of training epochs
    per_device_train_batch_size=4,  # batch size per device during training
    per_device_eval_batch_size=4,   # batch size for evaluation
    warmup_steps=50,                # number of warmup steps for learning rate scheduler
    weight_decay=0.01,               # strength of weight decay
    logging_dir='./logs',            # directory for storing logs
    logging_steps=10,
    fp16=True,                       # Enable mixed precision training
    gradient_accumulation_steps=4,   # Number of updates steps to accumulate before performing a backward/update pass
)

trainer = Trainer(
    model=model,                          # The model to be trained
    args=training_args,                   # Training arguments
    train_dataset=train_data_tokenized,  # Tokenized training dataset
    eval_dataset=eval_data_tokenized,    # Tokenized evaluation dataset
)

# Train the model
trainer.train()

# Evaluate the model
trainer.evaluate()
```

```python
from rouge_score import rouge_scorer
import torch

# Initialize a scorer
scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)

# Assuming eval_data is your DataFrame
# eval_data = pd.read_csv("your_test_dataset.csv")  # Load your DataFrame here

# Compute ROUGE scores
scores = []

# We will generate predictions and compute scores for each example individually
# to avoid GPU memory issues
for idx, row in test_data.iterrows():
    # Check if the row contains any NaN values or non-string values
    if pd.isna(row["Text"]) or pd.isna(row["Summary"]) or not isinstance(row["Text"], str) or not isinstance(row["Summary"], str):
        print(f"Skipping row {idx} due to NaN values or non-string values.")
        continue

    # Tokenize the input text and convert to Tensor
    inputs = tokenizer(row["Text"], padding=True, truncation=True, max_length=512, return_tensors="pt").to("cuda")

    # Generate prediction
    prediction = model.generate(**inputs, max_length=512)

    # Decode the prediction
    predicted_summary = tokenizer.decode(prediction[0], skip_special_tokens=True)

    # Compute score
    score = scorer.score(row["Summary"], predicted_summary)

    # Append the score to our list
    scores.append(score)

# Print the scores
for score in scores:
    print(score)
```
